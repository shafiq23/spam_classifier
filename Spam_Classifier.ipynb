# Spam classifier 

  In this notebook we are going to see how we can use NLP to detect spam messages and how to can classify them using machine learning libraries and techniques.

source : https://archive.ics.uci.edu/ml/datasets/sms+spam+collection

```python
#  Import dataset 

import pandas as pd

messages = pd.read_csv('drive/MyDrive/smsspamcollection/SMSSpamCollection', sep='\t', names=["label", "message"])
```

```python
messages.head()
```

```python
# Data Cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
corpus = []

for i in range(0, len(messages)):
  review = re.sub('[^a-zA-Z]', ' ', messages["message"][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus[:10]
```

```python
# Creating the bag of words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500) # setting max features value to limit the features
X = cv.fit_transform(corpus).toarray()
```

```python
X[:10]
```

```python
# Create dummies
y = pd.get_dummies(messages["label"])
y = y.iloc[:, 1].values # selecting spam column as reference

# Creating train and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train model using Navie Bayes calssifier
from sklearn.naive_bayes import MultinomialNB

spam_detect_model = MultinomialNB().fit(X_train, y_train)

y_preds = spam_detect_model.predict(X_test)
```

```python
y_preds
```

```python
# check with confusion matrix
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

values = confusion_matrix(y_test, y_preds)
values
```

```python
# Check the accuracy of the model
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_preds)
acc
```
